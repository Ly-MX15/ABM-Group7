# Analyze Sensitivity Analysis

```python
from src.SensitivityAnalysis import load_data, analyse, plot_indices

# Load the results
results = load_data(1)

# Create problem used in SALib
ranges = {
    "vision_mean": [1, 6],
    "metabolism_mean": [1, 6],
    "max_age_mean": [70, 100],
    "repopulate_factor": [5, 15],
    "cell_regeneration": [1, 5],
}

problem = {
    "num_vars": len(ranges),
    "names": list(ranges.keys()),
    "bounds": list(ranges.values())
}

# Perform the analysis
indices = analyse(problem, results)

# Plot the results
fig, ax = plot_indices(indices[0], indices[1], problem)

# Save the plot
fig.savefig("SensitivityAnalysis/split_1/indices.png", dpi=300, bbox_inches="tight")
```
